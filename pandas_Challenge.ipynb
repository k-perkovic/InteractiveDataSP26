### 3. Do you agree with Murray's claim that open-source tools that avoid proprietary software are best for publishing data visualization? In a few sentances, explain why you do or do not agree.
I agree with Murray's claim that open-source tools that avoid proprietary software are best for publishing data visualization. Murray states, "The more accessible your visualization, the greater your audience and your impact" as he believes that these tools are more impactful than proprietary software. These types of tools allow for more complex and larger data sets to be spread and viewable amongst different devices with different display requirements, leading to greater accessibility and visibility of the data. This is extremely important to consider as many users view data from a range of different devices.
