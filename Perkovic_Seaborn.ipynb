```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
spotify = pd.read_csv("spotify-2023.csv", encoding="latin1")
spotify.head()
```

```python
spotify.keys()
```

```python
hist1 = sns.histplot(data=spotify, x='bpm', kde='true')
hist1.plot()
```

```python
releasemonthgroup = spotify.groupby('released_month').count()['track_name']
releasemonthgroup
```

```python
releasemonthgroupvalues =  releasemonthgroup.values
releasemonthgroupvalues
```

```python
releasemonthindex = releasemonthgroup.index
releasemonthindex
```

```python
cat1 = sns.catplot(x= releasemonthindex, y= releasemonthgroupvalues)
plt.show()
```

```python
cat1 = sns.catplot(x= releasemonthindex, y= releasemonthgroupvalues, kind= "bar")
plt.show()
```

```python
cat1 = sns.catplot(x= releasemonthindex, y= releasemonthgroupvalues, kind="bar", hue = releasemonthindex)
plt.show()
```

```python
cat1 = sns.catplot(x= releasemonthindex, y= releasemonthgroupvalues, kind="bar", hue = releasemonthindex, palette="Spectral", legend = False)
plt.show()
```

```python
rel1 = sns.relplot(data=spotify, x="energy_%", y="bpm")
plt.show()
```

```python
modelist = spotify.groupby("mode").count()["track_name"]
modelist
```

```python
spotify['month_cat'] = spotify['released_month'].astype('category')
spotify['month_cat'] = spotify['month_cat'].cat.rename_categories(releasemonth_names)
```

```python
rel1 = sns.relplot(data=spotify, x="energy_%", y="bpm", hue="month_cat", palette=cp,
                   style="mode", markers=["8", ">"], size="month_cat")
plt.show()
```

```python
rel6 = sns.relplot(data=spotify, x="energy_%", y="bpm", col = "mode", hue = "mode", legend=False, alpha= 0.3)
plt.show()
```

```python
rel7 = sns.relplot(data=spotify, x="energy_%", y="bpm", col = "released_month", col_wrap=4)
plt.show()
```
