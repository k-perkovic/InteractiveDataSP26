A career goal of mine is being able to understand which platform holds the most social media engagement in collegiate esports. By using a stacked bars graph, viewers are able to visually see which platform reaches the most engagement.

```python
cces = pd.read_csv('cces_example_data.csv')
cces.head()
```

```python
fig1, axs1 = plt.subplots()
axs1.set_xlabel('Age Histogram')
axs1.set_ylabel('Political Identity')
```

```python
cces['Eval'] = cces['Evaluation'].astype('category')
```

```python
eval_groups = cces.groupby('Eval').count()['sex']
eval_groups
```

```python
fig2, ax2 = plt.subplots()
ax2.bar(cces['Eval'], cces['nonwhite'])
plt.show()
```

```python
fig3, ax3 = plt.subplots()
ax3.scatter(cces['ConMediaUse'], cces['ImmiPhobia'], color = 'lightgreen')
plt.show()
```

Since there is a large number of individual points in the scatterplot above, the readings from Wilke suggest to create a 2D histogram. Instead of individual points, a user should bin the data, subdivide the x and y plane into smaller rectangles, and then count and color the rectangles. Referring to the figure 18.6 that contains these visualizations, Wilke states that, "This visualization clearly highlights several important features" when referring to figure 18.6 in Chapter 18 - _Handling overlapping points_. This resulting in a more dense and understandable graph.
